```python
plt.figure(figsize=(8, 4))
sns.histplot(penguins['body_mass_g'], kde=True, bins=20, stat='density', color='skyblue')
plt.title('Distribution of body_mass_g')
plt.xlabel('body_mass_g')
plt.ylabel('Density')
plt.show()
```

```python


# Q-Q plot for 'body_mass_g' to see how it aligns with
plt.figure(figsize=(8, 4))
stats.probplot(penguins['body_mass_g'], dist="norm", plot=plt)
plt.title('Q-Q Plot for body_mass_g')
plt.show()
```

```python





body_mass_data = penguins['body_mass_g'].dropna()  # Remove missing values for plotting

distributions = [norm, lognorm, expon, gamma]
colors = ['blue', 'red', 'green', 'orange']

plt.figure(figsize=(8, 4))
sns.histplot(body_mass_data, kde=True, stat='density', color='skyblue', label='Histogram')

for distribution, color in zip(distributions, colors):
    frozen_dist = distribution(*distribution.fit(body_mass_data))
    x = sorted(body_mass_data)
    y = frozen_dist.pdf(x)
    plt.plot(x, y, label=f'{distribution.name}', color=color)

plt.title('Distribution of body_mass_g and Fitted Distributions')
plt.xlabel('body_mass_g')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm, lognorm, expon, gamma

penguins = sns.load_dataset("penguins")

body_mass_data = penguins['body_mass_g'].dropna()  # Remove missing values for plotting

# Theoretical distributions
distributions = [norm, lognorm, expon, gamma]
colors = ['blue', 'red', 'green', 'orange']

plt.figure(figsize=(8, 4))
sns.histplot(body_mass_data, kde=True, stat='density', color='skyblue', label='Histogram')

for distribution, color in zip(distributions, colors):
    frozen_dist = distribution(*distribution.fit(body_mass_data))
    x = sorted(body_mass_data)
    y = frozen_dist.pdf(x)
    plt.plot(x, y, label=f'{distribution.name}', color=color)

plt.yscale('log')

plt.title('Distribution of body_mass_g and Fitted Distributions')
plt.xlabel('body_mass_g')
plt.ylabel('Density (log scale)')
plt.legend()

#plot
plt.show()
```
