```python
#Libraries needed to do the analysis
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# initially in task 2 I loaded the url of the dataset as I - an idiot - didn't realise it was built-in within seaborn.
penguins = sns.load_dataset("penguins")
```

```python
# Get some information and visualisations of the data
print(penguins.head(5))
print(penguins.info())
print(penguins.describe())
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(penguins['body_mass_g'], kde=True, bins=20, stat='density', color='skyblue')
plt.title('Distribution of body_mass_g')
plt.xlabel('body_mass_g')
plt.ylabel('Density')
plt.show()
```

```python


# Q-Q plot for 'body_mass_g' to see how it aligns with
plt.figure(figsize=(8, 4))
stats.probplot(penguins['body_mass_g'], dist="norm", plot=plt)
plt.title('Q-Q Plot for body_mass_g')
plt.show()
```
